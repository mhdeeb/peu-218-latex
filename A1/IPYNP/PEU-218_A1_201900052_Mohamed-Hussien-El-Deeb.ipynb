```python
from matplotlib import colors
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker

%matplotlib widget
```

```python
should_save = True
dpi = 200
save_location = "../"

VEC_O = np.zeros(3)
```

## Question 1

```python
def question_1():
    fig = plt.figure()
    
    ax = fig.add_subplot(111, projection="3d")
    
    ax.set_xlim([-3, 3])
    ax.set_ylim([-3, 3])
    ax.set_zlim([-3, 3])

    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
    
    vec_a = np.array([1, 1, 0])
    vec_q = np.array([3, 0, 0])
    vec_p = np.array([1, 2, 3])

    ax.quiver(*VEC_O, *vec_a, color="r", label=r"$\overrightarrow{A}$")
    ax.quiver(*VEC_O, *vec_p, color="g", label=r"$\overrightarrow{P}$")
    ax.quiver(*VEC_O, *vec_q, color="b", label=r"$\overrightarrow{Q}$")
    ax.quiver(
        *VEC_O, *(vec_q - vec_p), color="#088F8F", label=r"$\overrightarrow{Q}-\overrightarrow{P}$"
    )
    ax.quiver(
        *vec_p,
        *(vec_q - vec_p),
        color="#088F8F",
        linestyle="--",
        label=r"$\overrightarrow{Q}-\overrightarrow{P}$",
    )

    plt.legend()
    
    if should_save:
        fig.savefig(f"{save_location}Q1.png", transparent=True, dpi=dpi, bbox_inches="tight")
```

```python
question_1()
```

## Question 2

```python
def question_2():
    fig = plt.figure()
    
    ax = fig.add_subplot(111, projection="3d")
    
    ax.set_xlim([-1.5, 1.5])
    ax.set_ylim([-1.5, 1.5])
    ax.set_zlim([-1.5, 1.5])

    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
    
    ax.view_init(17, 32)
    
    vec_p = np.array([3, 1, 2])
    vec_n_hat = np.array([1, -2, 1])/np.sqrt(6)
    mu = 2/np.sqrt(6)
    
    x_values = np.arange(-10, 10, 0.5)
    y_values = np.arange(-10, 10, 0.5)

    x, y = np.meshgrid(x_values, y_values)

    z = -x + 2 * y + 5

    ax.quiver(
        *VEC_O,
        *vec_p,
        color="r",
        linestyle="-",
        label=r"$\overrightarrow{P}$",
    )
    
    ax.quiver(
        *VEC_O,
        *(vec_p + mu * vec_n_hat),
        color="g",
        linestyle="-",
        label=r"$\overrightarrow{D}$",
    )
    
    ax.quiver(
        *vec_p,
        *(mu * vec_n_hat),
        color="b",
        linestyle="-",
        label=r"$\overrightarrow{P}-\overrightarrow{D}$",
    )
    
    ax.plot_surface(x, y, z, alpha=0.5, label="plane")

    plt.legend()
    
    if should_save:
        fig.savefig(f"{save_location}Q2.png", transparent=True, dpi=dpi, bbox_inches="tight")
```

```python
question_2()
```

## Question 3

```python
def question_3():
    fig = plt.figure()

    ax = fig.add_subplot(111, projection="3d")
     
    ax.set_xlim([-1.5, 1.5])
    ax.set_ylim([-1.5, 1.5])
    ax.set_zlim([-1.5, 1.5])

    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
    
    ax.view_init(14, -14)
    
    
    t = np.linspace(-10, 10, 100)
    
    x = 2*t - 1
    y = -t - 1
    z = t + 2
    
    vec_v = np.array([1, -1, 0])
    vec_mu = np.array([2, -1, 1])
    vec_x = np.array([1, 1, -1])
    
    ax.plot(x, y, z, label="line")
    
    ax.quiver(
        *VEC_O,
        *vec_v,
        color="r",
        linestyle="-",
        label=r"$\overrightarrow{v}$",
    )
    
    ax.quiver(
        *VEC_O,
        *vec_mu,
        color="g",
        linestyle="-",
        label=r"$\overrightarrow{\mu}$",
    )
    
    ax.quiver(
        *VEC_O,
        *vec_x,
        color="b",
        linestyle="-",
        label=r"$\overrightarrow{x}$",
    )
    
    plt.legend()
    
    if should_save:
        fig.savefig(f"{save_location}Q3.png", transparent=True, dpi=dpi, bbox_inches="tight")
```

```python
question_3()
```

## Question 4

```python
def question_4():
    fig = plt.figure()

    ax = fig.add_subplot(111, projection="3d")
     
    ax.set_xlim([-1.5, 1.5])
    ax.set_ylim([-1.5, 1.5])
    ax.set_zlim([-1.5, 1.5])

    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
    
    ax.view_init(5, -67)
    
    
    t = np.linspace(-10, 10, 100)
    
    x = 1 + t * 4
    y = 2 + t * 5
    z = 3 + t * 6
    
    ax.plot(x, y, z, label="line")
    
    lam = -7/31
    vec_a = np.array([1, 2, 3])
    vec_b = np.array([4, 5, 6])
    
    ax.quiver(
        *VEC_O,
        *(vec_a + lam * vec_b),
        color="r",
        linestyle="-",
        label=r"$\overrightarrow{v}$",
    )
    
    x_values = np.arange(-10, 10, 0.5)
    y_values = np.arange(-10, 10, 0.5)

    x, y = np.meshgrid(x_values, y_values)

    z = -2/3*x - 1/3*y + 2
    
    ax.plot_surface(x, y, z, alpha=0.5, label="plane")
    
    plt.legend()
    
    if should_save:
        fig.savefig(f"{save_location}Q4.png", transparent=True, dpi=dpi, bbox_inches="tight")
```

```python
question_4()
```
