```python
from matplotlib import colors
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker

%matplotlib widget
```

```python
def q1():
    O = np.array([0, 0, 0])
    A = np.array([1, 1, 0])
    Q = np.array([3, 0, 0])
    P = np.array([1, 2, 3])

    fig = plt.figure()
    ax = fig.add_subplot(111, projection="3d")
    ax.quiver(*O, *A, color="r", label=r"$\overrightarrow{A}$")
    ax.quiver(*O, *P, color="g", label=r"$\overrightarrow{P}$")
    ax.quiver(*O, *Q, color="b", label=r"$\overrightarrow{Q}$")
    ax.quiver(
        *O, *(Q - P), color="#088F8F", label=r"$\overrightarrow{Q}-\overrightarrow{P}$"
    )
    ax.quiver(
        *P,
        *(Q - P),
        color="#088F8F",
        linestyle="--",
        label=r"$\overrightarrow{Q}-\overrightarrow{P}$",
    )
    ax.set_xlim([-3, 3])
    ax.set_ylim([-3, 3])
    ax.set_zlim([-3, 3])

    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")

    plt.legend()
    
    fig.savefig("../Q1.png", transparent=True, dpi=100, bbox_inches="tight")
```

```python
q1()
```

```python
def q2():
    O = np.array([0, 0, 0])
    P = np.array([3, 1, 2])
    N = np.array([1, -2, 1])/np.sqrt(6)
    mu = 2/np.sqrt(6)
    
    x_values = np.arange(-10, 10, 0.5)
    y_values = np.arange(-10, 10, 0.5)

    X, Y = np.meshgrid(x_values, y_values)

    Z = -X + 2 * Y + 5

    fig = plt.figure()

    ax = fig.add_subplot(111, projection="3d")
    ax.quiver(
        *O,
        *P,
        color="r",
        linestyle="-",
        label=r"$\overrightarrow{P}$",
    )
    ax.quiver(
        *O,
        *(P + mu*N),
        color="g",
        linestyle="-",
        label=r"$\overrightarrow{D}$",
    )
    ax.quiver(
        *P,
        *(mu*N),
        color="b",
        linestyle="-",
        label=r"$\overrightarrow{P}-\overrightarrow{D}$",
    )
    ax.plot_surface(X, Y, Z, alpha=0.5)
    
    ax.set_xlim([-1.5, 1.5])
    ax.set_ylim([-1.5, 1.5])
    ax.set_zlim([-1.5, 1.5])
    
    ax.view_init(17, 32)

    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")

    plt.legend()
    fig.savefig("../Q2.png", transparent=True, dpi=100, bbox_inches="tight")
```

```python
q2()
```
