```python
from matplotlib import colors
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker

%matplotlib widget
```

```python
should_save = False
dpi = 50
save_location = "../"

VEC_O = np.zeros(3)
```

## Question 3

```python
def question_3():
    fig = plt.figure()
    
    ax = fig.add_subplot()
    
    ax.grid(color="black", linestyle="-", linewidth=0.5, which="major")
    ax.grid(color="grey", linestyle="--", linewidth=0.25, which="minor")

    ax.xaxis.set_major_locator(ticker.MultipleLocator(base=1))
    ax.yaxis.set_major_locator(ticker.MultipleLocator(base=1))
    ax.minorticks_on()
    
    ax.set_xlim([-6, 6])
    ax.set_ylim([-6, 6])

    ax.set_xlabel("X")
    ax.set_ylabel("Y")

    point1 = [0, -4]
    point2 = [-2, -4]
    x_values = [point1[0], point2[0]]
    y_values = [point1[1], point2[1]]
    ax.plot(x_values, y_values, linestyle="-", label="C1")
    
    x_values = np.linspace(-2, 2, 100)
    y_values = -x_values**2
    ax.plot(x_values, y_values, linestyle="-", label="C2")
    
    point1 = [2, -4]
    point2 = [5, 1]
    x_values = [point1[0], point2[0]]
    y_values = [point1[1], point2[1]]
    ax.plot(x_values, y_values, linestyle="-", label="C3")

    plt.legend()
    
    if should_save:
        fig.savefig(f"{save_location}Q3.png", transparent=True, dpi=dpi, bbox_inches="tight")
```

```python
question_3()
```
