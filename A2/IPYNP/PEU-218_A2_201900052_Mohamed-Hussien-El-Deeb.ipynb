```python
from matplotlib import colors
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker

%matplotlib widget
```

```python
should_save = False
dpi = 100
save_location = "../"

VEC_O = np.zeros(3)
```

## Question 3

```python
def question_3():
    fig = plt.figure()
    
    ax = fig.add_subplot()
    
    ax.grid(color="black", linestyle="-", linewidth=0.5, which="major")
    ax.grid(color="grey", linestyle="--", linewidth=0.25, which="minor")

    ax.xaxis.set_major_locator(ticker.MultipleLocator(base=1))
    ax.yaxis.set_major_locator(ticker.MultipleLocator(base=1))
    ax.minorticks_on()
    
    ax.set_xlim([-6, 6])
    ax.set_ylim([-6, 6])

    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    
    
    point1 = np.array([0, -4])
    point2 = np.array([-2, -4])
    
    ax.quiver(
        *point1,
        *(point2 - point1),
        label=r"C1",
        scale=1,
        angles='xy',
        scale_units="xy",
        color='#1f77b4',
        width=0.005
    )
    
    x_values = np.linspace(-2, 2, 100)
    y_values = -x_values**2
    ax.plot(x_values, y_values, color='#ff7f0e')
    
    point1 = np.array([x_values[-5], y_values[-5]])
    point2 = np.array([x_values[-1], y_values[-1]])
    
    ax.quiver(
        *point1,
        *(point2 - point1),
        label=r"C2",
        scale=1,
        angles='xy',
        scale_units="xy",
        color='#ff7f0e',
        width=0.005
    )
    
    point1 = np.array([2, -4])
    point2 = np.array([5, 1])
    
    ax.quiver(
        *point1,
        *(point2 - point1),
        label=r"C3",
        scale=1,
        angles='xy',
        scale_units="xy",
        color='#2ca02c',
        width=0.005
    )

    plt.legend()
    
    if should_save:
        fig.savefig(f"{save_location}Q3.png", transparent=True, dpi=dpi, bbox_inches="tight")
```

```python
question_3()
```
